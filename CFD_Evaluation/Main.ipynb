```python
import numpy as np
import matplotlib.pyplot as plt
import Aerodynamics
```

```python
# Get coordinates of airfoil
airfoil_coordinates = np.loadtxt('NACA0012.dat')
print(airfoil_coordinates)

# Plot the airfoil
plt.plot(airfoil_coordinates[:, 0], airfoil_coordinates[:, 1])
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.ylim(-.1, .1)
plt.title('NACA0012')
plt.show()
```

```python
# Create airfoil object to analyze properties
airfoil_name = 'my_airfoil'
airfoil = Aerodynamics.Airfoil(airfoil_coordinates, airfoil_name)

# Get lift-to-drag ratio
Reynolds_num = 1e6
airfoil.get_L_by_D(Reynolds_num)
```
