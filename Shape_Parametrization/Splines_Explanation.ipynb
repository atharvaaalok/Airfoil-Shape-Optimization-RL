```python
import numpy as np
import matplotlib.pyplot as plt
import Splines
```

```python
# Quadratic Bezier
cp0 = [(0, 0), (0, 2), (2, 0)]
cp1 = [(2, 0), (2, 4), (4, 2)]
cp2 = [(4, 2), (4, 6), (6, 4)]

cp_list = [cp0, cp1, cp2]
spline = Splines.Bezier_spline(cp_list)

n = 40
points = spline.generate_points(n)
x = points[:, 0]
y = points[:, 1]

plt.plot(x, y, marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.title('Quadratic Bezier Spline')
plt.show()
```

```python
# Cubic Bezier
cp0 = [(0, 0), (0, 2), (2, 0), (2, 2)]
cp1 = [(2, 2), (2, 4), (4, 2), (4, 4)]
cp2 = [(4, 4), (4, 6), (6, 4), (6, 6)]

cp_list = [cp0, cp1, cp2]
spline = Splines.Bezier_spline(cp_list)

n = 40
points = spline.generate_points(n)
x = points[:, 0]
y = points[:, 1]

plt.plot(x, y, marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.title('Cubic Bezier Spline')
plt.show()
```

```python
# Airfoil - Using Cubic Bezier Spline
cp0 = [[372, -111], [199, -20], [134, -27], [107, -27]]
cp1 = [[107, -27], [33, -27], [26, -60], [29, -94]]
cp2 = [[29, -94], [89, -139], [99, -55], [372, -111]]

cp_list = [cp0, cp1, cp2]
spline = Splines.Bezier_spline(cp_list)

n = 40
points = spline.generate_points(n)
x = points[:, 0]
y = points[:, 1]

plt.plot(x, y, marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.title('Airfoil: Cubic Bezier')
plt.show()
```
