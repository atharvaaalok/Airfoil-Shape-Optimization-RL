```python
import numpy as np
import matplotlib.pyplot as plt
import Curves
```

```python
p0 = (0, 0)
p1 = (0, 2)
p2 = (2, 0)
p3 = (2, 2)
```

```python
# Quadratic Bezier curve
quad_bezier = Curves.Quadratic_Bezier(p0, p1, p2)

# Generate points
n = 20
points = quad_bezier.generate_points(n)
x = points[:, 0]
y = points[:, 1]

# Plot the points
plt.plot(x, y, marker = 'o')
plt.title('Quadratic Bezier')
plt.show()
```

```python
# Cubic Bezier curve
cubic_bezier = Curves.Cubic_Bezier(p0, p1, p2, p3)

# Generate points
n = 20
points = cubic_bezier.generate_points(n)
x = points[:, 0]
y = points[:, 1]

# Plot the points
plt.plot(x, y, marker = 'o')
plt.title('Cubic Bezier')
plt.show()
```

```python
# Catmull Rom curve
catmull_rom = Curves.CatmullRom_curve(p0, p1, p2, p3)

# Generate points
n = 20
points = catmull_rom.generate_points(n)
x = points[:, 0]
y = points[:, 1]

# Plot the points
plt.plot(x, y, marker = 'o')
plt.title('Catmull Rom')
plt.show()
```
