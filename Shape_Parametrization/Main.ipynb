```python
import numpy as np
import matplotlib.pyplot as plt
import Curves
from Splines import Bezier_spline, B_spline
```

```python
# Quadratic Bezier
cp0 = [(0, 0), (0, 2), (2, 0)]
cp1 = [(2, 0), (2, 4), (4, 2)]
cp2 = [(4, 2), (4, 6), (6, 4)]

cp_list = [cp0, cp1, cp2]
spline = Bezier_spline(cp_list)

n = 40
points = spline.generate_points(n)
x = points[:, 0]
y = points[:, 1]

plt.plot(x, y, marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Cubic Bezier
cp0 = [(0, 0), (0, 2), (2, 0), (2, 2)]
cp1 = [(2, 2), (2, 4), (4, 2), (4, 4)]
cp2 = [(4, 4), (4, 6), (6, 4), (6, 6)]

cp_list = [cp0, cp1, cp2]
a = Bezier_spline(cp_list)

n = 40
my_points = a.generate_points(n)
x = my_points[:, 0]
y = my_points[:, 1]

plt.plot(x, y, marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Airfoil
cp0 = [[372, -111], [199, -20], [134, -27], [107, -27]]
cp1 = [[107, -27], [33, -27], [26, -60], [29, -94]]
cp2 = [[29, -94], [89, -139], [99, -55], [372, -111]]

cp_list = [cp0, cp1, cp2]
a = Bezier_spline(cp_list)

n = 40
my_points = a.generate_points(n)
x = my_points[:, 0]
y = my_points[:, 1]

plt.plot(x, y, marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Airfoil using B-Splines
n = 40

x_cp = [0, 0, 2, 2]
y_cp = [0, 2, 0, 2]
cp0 = [(x_cp[i], y_cp[i]) for i in range(4)]

from Curves import B_Spline_curve

a = B_Spline_curve(*cp0)

my_points = a.generate_points(n)
x = my_points[:, 0]
y = my_points[:, 1]

plt.plot(x, y, marker = 'o')
plt.plot(x_cp, y_cp, linestyle = 'None', marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

# Airfoil using B-Splines
x_cp = [0, 2, 2, 2]
y_cp = [2, 0, 2, 3]
cp1 = [(x_cp[i], y_cp[i]) for i in range(4)]

from Curves import B_Spline_curve

b = B_Spline_curve(*cp1)

my_points = b.generate_points(n)
x = my_points[:, 0]
y = my_points[:, 1]

plt.plot(x, y, marker = 'o')
plt.plot(x_cp, y_cp, linestyle = 'None', marker = 'o')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```
